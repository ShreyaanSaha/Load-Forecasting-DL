```python
import numpy as np

# Z-score method
from scipy import stats

z_scores = np.abs(stats.zscore(X_train_scaled))
outliers_z = np.where(z_scores > 3)  # Z-score threshold can be 2.5 or 3
print("Outliers detected using Z-score method:", outliers_z)

# IQR method
Q1 = np.percentile(X_train_scaled, 25, axis=0)
Q3 = np.percentile(X_train_scaled, 75, axis=0)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_iqr = np.where((X_train_scaled < lower_bound) | (X_train_scaled > upper_bound))
print("Outliers detected using IQR method:", outliers_iqr)
```

```python
# Clip using IQR bounds
X_train_scaled_clipped = np.clip(X_train_scaled, lower_bound, upper_bound)
```

# **SIMPLE RNN Architechture**

```python
from tensorflow.keras import layers, regularizers, Sequential
from tensorflow.keras.optimizers import Adam
import numpy as np

# Assuming X_train_scaled_clipped and X_test_scaled_clipped are numpy arrays
# Reshape data to be 3D: (samples, time steps, features)
if len(X_train_scaled_clipped.shape) == 2:
    X_train_scaled_clipped = np.expand_dims(X_train_scaled_clipped, axis=1)
if len(X_train_scaled_clipped.shape) > 3:
    X_train_scaled_clipped = X_train_scaled_clipped.reshape(-1, 1, 23)

if len(X_test_scaled_clipped.shape) == 2:
    X_test_scaled_clipped = np.expand_dims(X_test_scaled_clipped, axis=1)
if len(X_test_scaled_clipped.shape) > 3:
    X_test_scaled_clipped = X_test_scaled_clipped.reshape(-1, 1, 23)

hidden_units = 100

model = keras.Sequential([
    keras.layers.SimpleRNN(hidden_units, activation='tanh', input_shape=(1, 23), 
                           kernel_initializer='he_normal', kernel_regularizer=keras.regularizers.l2(0.001), return_sequences=True),
    
    keras.layers.SimpleRNN(hidden_units, activation='tanh',
                           kernel_initializer='he_normal', kernel_regularizer=keras.regularizers.l2(0.001), return_sequences=True),
    
    keras.layers.SimpleRNN(hidden_units, activation='tanh',
                           kernel_initializer='he_normal', kernel_regularizer=keras.regularizers.l2(0.001), return_sequences=True),
    
    keras.layers.SimpleRNN(hidden_units, activation='tanh',
                           kernel_initializer='he_normal', kernel_regularizer=keras.regularizers.l2(0.001)),
    
    keras.layers.Dense(1, activation='linear')
])

model.compile(optimizer='adam', loss='mae')
```

```python
# Fit the model
history = model.fit(X_train_scaled_clipped, y_train, 
                    validation_data=(X_test_scaled_clipped, y_test), 
                    epochs=30, batch_size=32)

# Evaluate the model
loss = model.evaluate(X_test_scaled_clipped, y_test)
print("Test MAE loss:", loss)
```

 # **Combined Stacked and Bidirectional LSTM Model:**

```python
from tensorflow import keras

hidden_units = 100

model = keras.Sequential([
    keras.layers.Bidirectional(keras.layers.LSTM(hidden_units, activation='selu', input_shape=(1, 23), 
                                                 kernel_initializer='he_normal', kernel_regularizer=keras.regularizers.l2(0.001), 
                                                 return_sequences=True)),
    
    keras.layers.Bidirectional(keras.layers.LSTM(hidden_units, activation='selu', 
                                                 kernel_initializer='he_normal', kernel_regularizer=keras.regularizers.l2(0.001), 
                                                 return_sequences=True)),
   
    
    keras.layers.Bidirectional(keras.layers.LSTM(hidden_units, activation='selu', 
                                                 kernel_initializer='he_normal', kernel_regularizer=keras.regularizers.l2(0.001), 
                                                 return_sequences=True)),
    keras.layers.Bidirectional(keras.layers.LSTM(hidden_units, activation='selu', 
                                                 kernel_initializer='he_normal', kernel_regularizer=keras.regularizers.l2(0.001))),
  
    keras.layers.Dense(1, activation='linear')
])

model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.001), loss='mae')
```

```python
# Fit the model
history = model.fit(X_train_scaled_clipped, y_train, 
                    validation_data=(X_test_scaled_clipped, y_test), 
                    epochs=100, batch_size=32)

# Evaluate the model
loss = model.evaluate(X_test_scaled_clipped, y_test)
print("Test MAE loss:", loss)
```

# **MAE,MSE,RMSE,R-squared Score Calculation:**

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Assuming model is already trained and X_test_scaled_clipped, y_test are defined
y_pred = model.predict(X_test_scaled_clipped)

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error (MAE):", mae)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error (MSE):", mse)

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)
print("Root Mean Squared Error (RMSE):", rmse)

# Calculate R-squared (R2) score
r2 = r2_score(y_test, y_pred)
print("R-squared (R2) score:", r2*100)
```
